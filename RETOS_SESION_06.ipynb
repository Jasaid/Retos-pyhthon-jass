```python
# LIMPIEZA DE DATOS

# Eliminar duplicados
df = df_excel.drop_duplicates()

# Dejar s√≥lo valores NO Nulos
df = df_excel.dropna()

# CREAR Y ANALIZAR SERIES
Valor_CLP = df["Valor_CLP"]  # Precio en Pesos Chilenos
Total_Superficie_M2 = df["Total_Superficie_M2"]  # Metros_cuadrados

# C√°lculo de estad√≠sticas b√°sicas
media_precio = Valor_CLP.mean()
mediana_precio = Valor_CLP.median()
print(f"Media del precio: {media_precio}")
print(f"Mediana del precio: {mediana_precio}")
media_superficie = Total_Superficie_M2.mean()
mediana_superficie = Total_Superficie_M2.median()
print(f"Media de la superficie: {media_superficie}")
print(f"Mediana de la superficie: {mediana_superficie}")
viviendas_caras = Valor_CLP > media_precio
viviendas_caras = df[viviendas_caras]
print(viviendas_caras)

print('____________________________________________________________')

# A√ëADIR Y MODIFICAR COLUMNAS
# Agregar una nueva columna con el precio por metro cuadrado
df["precio_por_m2"] = df["Valor_CLP"] / df["Total_Superficie_M2"]

# Calcular la mediana del precio por metro cuadrado
mediana_precio_por_m2 = df["precio_por_m2"].median()
print(f"Mediana del precio por metro cuadrado: {mediana_precio_por_m2}")
print(df.head())

# GENERAR INFORME DE HALLAZGOS
informe = f"""
An√°lisis de precios de viviendas:
- Media de los precios: {media_precio}
- Mediana de los precios: {mediana_precio}
- Total de viviendas analizadas: {len(df)}
- Viviendas con precios superiores a la media: {len(viviendas_caras)}
"""
print(informe)
```
