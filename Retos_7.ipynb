📊 Calcular la suma total de las calificaciones de cada estudiante.

Puedes usar un filtro para obtener las calificaciones de cada estudiante: df[df['Estudiante'] == 'NOMBRE']['Calificación'].FUNCION()
Este proceso lo veremos mas a detalle en la siguiente sesión.
🔢 Calcular el promedio de las calificaciones de cada estudiante.

📏 Determinar la calificación más alta y más baja registrada en el DataFrame.

📄 Mostrar los resultados, e imprime un mensaje similar:

✉️ Suma y promedio de calificaciones de Ana: 253 y 84.33
✉️ Suma y promedio de calificaciones de Luis: 265 y 88.33
✉️ Suma y promedio de calificaciones de María: 263 y 87.67
✉️ Calificación más alta en el DataFrame: 92
✉️ Calificación más baja en el DataFrame: 78

```python
import pandas as pd
```

```python
# Crear data frame

data = {
    'Estudiante': ['Ana', 'Ana', 'Ana', 'Luis', 'Luis', 'Luis', 'María', 'María', 'María'],
    'Materia': ['Matemáticas', 'Historia', 'Ciencias', 'Matemáticas', 'Historia', 'Ciencias', 'Matemáticas', 'Historia', 'Ciencias'],
    'Calificación': [85, 90, 78, 88, 85, 92, 90, 88, 85]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Calcular la suma total de las calificaciones de cada estudiante.
calificaciones_ana = df[df['Estudiante'] == 'Ana']['Calificación'].sum()
calificaciones_luis = df[df['Estudiante'] == 'Luis']['Calificación'].sum()
calificaciones_maria = df[df['Estudiante'] == 'María']['Calificación'].sum()
print(f"Suma de calificaciones de Ana: {calificaciones_ana}")
print(f"Suma de calificaciones de Luis: {calificaciones_luis}")
print(f"Suma de calificaciones de María: {calificaciones_maria}")
```

```python
# Calcular el promedio de las calificaciones de cada estudiante.
promedio_ana = df[df['Estudiante'] == 'Ana']['Calificación'].mean()
promedio_luis = df[df['Estudiante'] == 'Luis']['Calificación'].mean()
promedio_maria = df[df['Estudiante'] == 'María']['Calificación'].mean()
print(f"Promedio de calificaciones de Ana: {promedio_ana}")
print(f"Promedio de calificaciones de Luis: {promedio_luis}")
print(f"Promedio de calificaciones de María: {promedio_maria}")
```

```python
# Determinar la calificación más alta y más baja registrada en el DataFrame.
calificacion_max = df['Calificación'].max()
calificacion_min = df['Calificación'].min()
print(f"Calificación más alta en el DataFrame: {calificacion_max}")
print(f"Calificación más baja en el DataFrame: {calificacion_min}")
```

```python
# Mostrar los resultados, e imprime un mensaje similar:
# ✉️ Suma y promedio de calificaciones de Ana: 253 y 84.33
# ✉️ Suma y promedio de calificaciones de Luis: 265 y 88.33
# ✉️ Suma y promedio de calificaciones de María: 263 y 87.67
# ✉️ Calificación más alta en el DataFrame: 92
# ✉️ Calificación más baja en el DataFrame: 78
print(f"Suma y promedio de calificaciones de Ana: {calificaciones_ana} y {promedio_ana}")
print(f"Suma y promedio de calificaciones de Luis: {calificaciones_luis} y {promedio_luis}")
print(f"Suma y promedio de calificaciones de María: {calificaciones_maria} y {promedio_maria}")
print(f"Calificación más alta en el DataFrame: {calificacion_max}")
print(f"Calificación más baja en el DataFrame: {calificacion_min}")
```

```python
import pandas as pd
import numpy as np
```

```python
# Crear data frame

data = {
    'ID_Cliente': [101, 102, 103, 104, 105, 106],
    'Fecha_Encuesta': ['2023-01-01', '2023-01-02', np.nan, '2023-01-04', '2023-01-05', '2023-01-06'],
    'Puntuacion_Satisfaccion': [5, 4, 3, np.nan, 5, 4],  # Escala de 1 a 5
    'Comentarios': ['Todo excelente', np.nan, 'Buen servicio', 'Regular', np.nan, 'Muy bueno']
}

df = pd.DataFrame(data)
df.head(10)
```

```python
# Imprime el numero de NaNs, para filas y columnas:
# Conteo de valores NaN por columna:
print("Conteo de valores NaN por columna:")
print(df.isna().sum())
```

```python
# Eliminar filas donde 'Fecha_Encuesta'o 'Puntuacion_Satisfaccion' contienen por lo menos un valor nulo.
# Usa el método .copy() para manipulaciones seguras.
# 📝 Manejo de comentarios faltantes:

df_filtrado = df.dropna(subset=['Fecha_Encuesta', 'Puntuacion_Satisfaccion']).copy()
df_filtrado['Comentarios'] = df_filtrado['Comentarios'].fillna('Sin comentarios')
df_filtrado = df_filtrado.reset_index(drop=True)
df_filtrado = df_filtrado.rename(columns={'Puntuacion_Satisfaccion': 'Calificacion'})
df_filtrado.head(10)
```

```python
#Reorganiza los índices después de eliminar filas.
# ✏️ Estandarización de nombres de columnas:
df_filtrado.columns = df_filtrado.columns.str.lower().str.replace(' ', '_')
df_filtrado.head(10)
print(df_filtrado.columns)
```

```python
# Cambia el nombre de la columna 'puntuacion_satisfaccion' a 'calificacion' para reflejar mejor los datos.
# 🖨️ Mostrar el DataFrame Actualizado:
df.rename(columns={'puntuacion_satisfaccion': 'calificacion'}, inplace=True)
print(df)
```
