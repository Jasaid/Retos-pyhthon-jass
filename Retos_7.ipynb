üî¢ Calcular el promedio de las calificaciones de cada estudiante.

üìÑ Mostrar los resultados, e imprime un mensaje similar:

‚úâÔ∏è Suma y promedio de calificaciones de Ana: 253 y 84.33

‚úâÔ∏è Suma y promedio de calificaciones de Mar√≠a: 263 y 87.67

```python
import pandas as pd
```

```python
# Crear data frame

data = {
    'Estudiante': ['Ana', 'Ana', 'Ana', 'Luis', 'Luis', 'Luis', 'Mar√≠a', 'Mar√≠a', 'Mar√≠a'],
    'Materia': ['Matem√°ticas', 'Historia', 'Ciencias', 'Matem√°ticas', 'Historia', 'Ciencias', 'Matem√°ticas', 'Historia', 'Ciencias'],
    'Calificaci√≥n': [85, 90, 78, 88, 85, 92, 90, 88, 85]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Calcular la suma total de las calificaciones de cada estudiante.
calificaciones_ana = df[df['Estudiante'] == 'Ana']['Calificaci√≥n'].sum()
calificaciones_luis = df[df['Estudiante'] == 'Luis']['Calificaci√≥n'].sum()
calificaciones_maria = df[df['Estudiante'] == 'Mar√≠a']['Calificaci√≥n'].sum()
print(f"Suma de calificaciones de Ana: {calificaciones_ana}")
print(f"Suma de calificaciones de Luis: {calificaciones_luis}")
print(f"Suma de calificaciones de Mar√≠a: {calificaciones_maria}")
```

```python
# Calcular el promedio de las calificaciones de cada estudiante.
promedio_ana = df[df['Estudiante'] == 'Ana']['Calificaci√≥n'].mean()
promedio_luis = df[df['Estudiante'] == 'Luis']['Calificaci√≥n'].mean()
promedio_maria = df[df['Estudiante'] == 'Mar√≠a']['Calificaci√≥n'].mean()
print(f"Promedio de calificaciones de Ana: {promedio_ana}")
print(f"Promedio de calificaciones de Luis: {promedio_luis}")
print(f"Promedio de calificaciones de Mar√≠a: {promedio_maria}")
```

```python
# Determinar la calificaci√≥n m√°s alta y m√°s baja registrada en el DataFrame.
calificacion_max = df['Calificaci√≥n'].max()
calificacion_min = df['Calificaci√≥n'].min()
print(f"Calificaci√≥n m√°s alta en el DataFrame: {calificacion_max}")
print(f"Calificaci√≥n m√°s baja en el DataFrame: {calificacion_min}")
```

```python
# Mostrar los resultados, e imprime un mensaje similar:
# ‚úâÔ∏è Suma y promedio de calificaciones de Ana: 253 y 84.33
# ‚úâÔ∏è Suma y promedio de calificaciones de Luis: 265 y 88.33
# ‚úâÔ∏è Suma y promedio de calificaciones de Mar√≠a: 263 y 87.67
# ‚úâÔ∏è Calificaci√≥n m√°s alta en el DataFrame: 92
# ‚úâÔ∏è Calificaci√≥n m√°s baja en el DataFrame: 78
print(f"Suma y promedio de calificaciones de Ana: {calificaciones_ana} y {promedio_ana}")
print(f"Suma y promedio de calificaciones de Luis: {calificaciones_luis} y {promedio_luis}")
print(f"Suma y promedio de calificaciones de Mar√≠a: {calificaciones_maria} y {promedio_maria}")
print(f"Calificaci√≥n m√°s alta en el DataFrame: {calificacion_max}")
print(f"Calificaci√≥n m√°s baja en el DataFrame: {calificacion_min}")
```

```python
import pandas as pd
import numpy as np
```

```python
# Crear data frame

data = {
    'ID_Cliente': [101, 102, 103, 104, 105, 106],
    'Fecha_Encuesta': ['2023-01-01', '2023-01-02', np.nan, '2023-01-04', '2023-01-05', '2023-01-06'],
    'Puntuacion_Satisfaccion': [5, 4, 3, np.nan, 5, 4],  # Escala de 1 a 5
    'Comentarios': ['Todo excelente', np.nan, 'Buen servicio', 'Regular', np.nan, 'Muy bueno']
}

df = pd.DataFrame(data)
df.head(10)
```

```python
# Imprime el numero de NaNs, para filas y columnas:
# Conteo de valores NaN por columna:
print("Conteo de valores NaN por columna:")
print(df.isna().sum())
```

```python
# Eliminar filas donde 'Fecha_Encuesta'o 'Puntuacion_Satisfaccion' contienen por lo menos un valor nulo.
# Usa el m√©todo .copy() para manipulaciones seguras.
# üìù Manejo de comentarios faltantes:

df_filtrado = df.dropna(subset=['Fecha_Encuesta', 'Puntuacion_Satisfaccion']).copy()
df_filtrado['Comentarios'] = df_filtrado['Comentarios'].fillna('Sin comentarios')
df_filtrado = df_filtrado.reset_index(drop=True)
df_filtrado = df_filtrado.rename(columns={'Puntuacion_Satisfaccion': 'Calificacion'})
df_filtrado.head(10)
```

```python
#Reorganiza los √≠ndices despu√©s de eliminar filas.
# ‚úèÔ∏è Estandarizaci√≥n de nombres de columnas:
df_filtrado.columns = df_filtrado.columns.str.lower().str.replace(' ', '_')
df_filtrado.head(10)
print(df_filtrado.columns)
```

```python
# Cambia el nombre de la columna 'puntuacion_satisfaccion' a 'calificacion' para reflejar mejor los datos.
# üñ®Ô∏è Mostrar el DataFrame Actualizado:
df.rename(columns={'puntuacion_satisfaccion': 'calificacion'}, inplace=True)
print(df)
```
